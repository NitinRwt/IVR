```python
import spacy
import csv
import pyttsx3

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Function to process a query
def process_query(query):
    doc = nlp(query)
    # Add your logic here to handle and analyze the processed query
    return doc

# Function to speak the response
def speak_response(response):
    engine = pyttsx3.init()
    engine.say(response)
    engine.runAndWait()

# Load the healthcare dataset
dataset_path = "healthcare_dataset.csv"

queries = []
responses = []

with open(dataset_path, "r", encoding="utf-8") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        queries.append(row["Query"])
        responses.append(row["Response"])

# Example usage:
for i, query in enumerate(queries):
    processed_query = process_query(query)
    response = responses[i]
    print(f"Query {i + 1}: {processed_query}")
    
    # You can add more sophisticated logic to generate a response based on the processed query
    # For now, we'll use the predefined responses from the dataset
    print(f"Response: {response}")

    # Speak out the response
    speak_response(response)
```
